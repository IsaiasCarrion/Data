### üîπ Ejercicio 13: Intercambio de Variables (Swapping)
**Nivel:** Intermedio

**Objetivo:** Intercambiar los valores de dos variables (un uso com√∫n de la asignaci√≥n m√∫ltiple).

**Consigna:**
1. Crea `x = 10` e `y = 20`.
2. Sin usar una tercera variable temporal (como `temp`), intercambia los valores de `x` e `y`.
3. Imprime `x` (deber√≠a ser 20) e `y` (deber√≠a ser 10).

```python
# Tu c√≥digo aqu√≠
x = 10
y = 20

# ... realiza el intercambio aqu√≠ ...

print(f"x ahora es: {x}")
print(f"y ahora es: {y}")
```

```python
# --- SOLUCI√ìN (mostrar solo si el usuario lo pide) ---
x = 10
y = 20

print(f"Valores iniciales: x={x}, y={y}")

# Intercambio (Swapping) usando asignaci√≥n m√∫ltiple
x, y = y, x

print(f"Valores intercambiados: x={x}, y={y}")
```

**Explicaci√≥n:**
* `x, y = y, x` es la forma "Pyth√≥nica" de intercambiar variables.
* Python primero eval√∫a la expresi√≥n de la derecha: `(y, x)` que es `(20, 10)`.
* Luego, asigna (desempaqueta) esa tupla a las variables de la izquierda `x, y`.
* En otros lenguajes, esto requerir√≠a una variable temporal: `temp = x; x = y; y = temp`.

### üîπ Ejercicio 16: Divisi√≥n Entera y M√≥dulo
**Nivel:** Intermedio

**Objetivo:** Entender los operadores `//` (divisi√≥n entera) y `%` (m√≥dulo/resto).

**Consigna:**
Quieres repartir 17 caramelos entre 5 ni√±os.
1. Crea `caramelos = 17` y `ni√±os = 5`.
2. Calcula cu√°ntos caramelos enteros recibe cada ni√±o (divisi√≥n entera `//`) y gu√°rdalo en `caramelos_por_ni√±o`.
3. Calcula cu√°ntos caramelos sobran (m√≥dulo `%`) y gu√°rdalo en `caramelos_sobrantes`.

```python
# Tu c√≥digo aqu√≠
caramelos = 17
ni√±os = 5
```

```python
# --- SOLUCI√ìN (mostrar solo si el usuario lo pide) ---

caramelos = 17
ni√±os = 5

# Divisi√≥n Entera: ¬øCu√°ntas veces cabe 5 en 17?
caramelos_por_ni√±o = caramelos // ni√±os

# M√≥dulo: ¬øCu√°nto sobra de la divisi√≥n 17 / 5?
caramelos_sobrantes = caramelos % ni√±os

print(f"Caramelos por ni√±o: {caramelos_por_ni√±o}")  # Deber√≠a ser 3
print(f"Caramelos sobrantes: {caramelos_sobrantes}")  # Deber√≠a ser 2
```

**Explicaci√≥n:**
* `//` (Divisi√≥n Entera): Descarta la parte decimal de la divisi√≥n. `17 / 5` es `3.4`, por lo que `17 // 5` es `3`.
* `%` (M√≥dulo): Devuelve el resto de la divisi√≥n. `5 * 3` es `15`. `17 - 15` es `2`.

```python
# Tu c√≥digo aqu√≠


def mi_funcion():
    # ...
    pass


# ...

# ... (Intenta imprimir aqu√≠)
```

```python
# --- SOLUCI√ìN (mostrar solo si el usuario lo pide) ---


def mi_funcion():
    variable_local = "Soy local"
    print(f"Dentro de la funci√≥n: {variable_local}")


# Llamamos a la funci√≥n
mi_funcion()

# Este intento dar√° un error
try:
    print(variable_local)
except NameError as e:
    print(f"\nError al intentar acceder fuera: {e}")
```

### üîπ Ejercicio 19: Modificando Global (Keyword `global`)
**Nivel:** Avanzado

**Objetivo:** Aprender a *modificar* una variable global desde una funci√≥n usando la palabra clave `global`.

**Consigna:**
1. Crea una variable global `contador_global = 100`.

3. (Pista: Si solo haces `contador_global += 1` dentro de la funci√≥n, dar√° un `UnboundLocalError`).
4. Usa la palabra clave `global` para que funcione.
5. Llama a `incrementar()` e imprime `contador_global` (deber√≠a ser 101).

```python
# Tu c√≥digo aqu√≠

contador_global = 100


def incrementar():
    # ...
    pass


incrementar()
print(contador_global)
```

```python
# --- SOLUCI√ìN (mostrar solo si el usuario lo pide) ---

contador_global = 100


def incrementar():
    # Sin la siguiente l√≠nea, Python crear√≠a una NUEVA variable local
    # 'contador_global' al intentar asignarle un valor,
    # pero fallar√≠a al intentar leerla (en el '+ 1').
    global contador_global

    contador_global += 1


incrementar()
print(f"El contador ahora es: {contador_global}")
```

**Explicaci√≥n:**
Si solo lees una variable global, Python la encuentra (Ejercicio 18). Pero si intentas *asignarle* un valor (con `=` o `+=`), Python asume que est√°s creando una *nueva variable local* con ese nombre.
`global contador_global` le dice a Python: "En esta funci√≥n, cuando digo `contador_global`, me refiero a la que est√° en el √°mbito global, no crees una nueva local".

### üîπ Ejercicio 20: Variables como Referencias (Mutabilidad)
**Nivel:** Avanzado

**Objetivo:** Comprender que las variables son "etiquetas" que apuntan a objetos en memoria, y la diferencia entre objetos mutables e inmutables.

**Consigna:**
1. Crea una lista `lista_a = [1, 2, 3]`.
2. Crea una variable `lista_b` asign√°ndole `lista_a` (Ej: `lista_b = lista_a`).
3. Usando `lista_b`, a√±ade el n√∫mero 4 al final (Ej: `lista_b.append(4)`).
4. Imprime `lista_a`.
5. (Opcional) Imprime `id(lista_a)` e `id(lista_b)` para ver sus direcciones de memoria.

**Pregunta (reflexi√≥n):** ¬øPor qu√© cambi√≥ `lista_a` si solo modificaste `lista_b`?

```python
# Tu c√≥digo aqu√≠
```

```python
# --- SOLUCI√ìN (mostrar solo si el usuario lo pide) ---

lista_a = [1, 2, 3]
lista_b = lista_a

print(f"ID de lista_a: {id(lista_a)}")
print(f"ID de lista_b: {id(lista_b)}")
print(f"¬øSon el mismo objeto? {lista_a is lista_b}")

print("\nModificando lista_b...")
lista_b.append(4)

print(f"Lista A (despu√©s): {lista_a}")
print(f"Lista B (despu√©s): {lista_b}")
```

**Explicaci√≥n:**
En Python, la asignaci√≥n (`=`) no copia el objeto, **copia la referencia** (la "etiqueta" o "puntero").

1.  `lista_a = [1, 2, 3]`: Python crea un objeto (la lista) en la memoria y la etiqueta `lista_a` apunta a √©l.
2.  `lista_b = lista_a`: Python **no** crea una nueva lista. Simplemente crea una nueva etiqueta `lista_b` y la hace apuntar **al mismo objeto** al que apunta `lista_a`.
3.  `lista_b.append(4)`: Como `lista_b` apunta al objeto original, este m√©todo modifica ese objeto. Las listas son **mutables** (pueden cambiar).
4.  `print(lista_a)`: Muestra `[1, 2, 3, 4]`, porque `lista_a` sigue apuntando a ese mismo objeto, que ahora ha sido modificado.

**Diferencia con Inmutables:**
Si hicieras esto con un `int` (que es **inmutable**):
`a = 5`
`b = a`
`b = b + 1` (o `b += 1`)
Aqu√≠, `a` seguir√≠a valiendo `5` y `b` valdr√≠a `6`. ¬øPor qu√©? Porque `b = b + 1` *no modifica* el objeto `5`. Crea un *nuevo* objeto `6` y hace que `b` apunte a √©l, mientras `a` sigue apuntando al objeto `5` original.
